## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import datetime as dt

Build=pd.read_csv("Building-Permits-2012-2022-residential.csv")
Occ=pd.read_csv("Occupancy-Permits-2012-2022-residential.csv")
Demo=pd.read_csv("Demolition-Permits-2012-2022-all.csv")
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
#Building
sns.pairplot(Build, x_vars=['TOTALNUMBER', 'TOTALVALUE'],
                   y_vars='YEAR', height=4, aspect=1, kind='scatter')
```

```python
#Occupancy
sns.pairplot(Occ, x_vars=['TOTALNUMBER', 'TOTALVALUE'],
                   y_vars='YEAR', height=4, aspect=1, kind='scatter')
```

```python
#Demolition
sns.pairplot(Demo, x_vars=['TOTALNUMBER', 'TOTALVALUE'],
                   y_vars='YEAR', height=4, aspect=1, kind='scatter')
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
#Assign x and y variables

X = Build['TOTALNUMBER']
y = Build['TOTALVALUE']

X2 = Occ['TOTALNUMBER']
y2 = Occ['TOTALVALUE']

X3 = Demo['TOTALNUMBER']
y3 = Demo['TOTALVALUE']
```

```python
#Building
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                   test_size = 0.3, random_state = 100)
```

```python
import statsmodels.api as sm


#Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
```

```python
lr = sm.OLS(y_train, X_train_sm).fit()

#Printing parameters
lr.params
```

```python
#Occupancy
from sklearn.model_selection import train_test_split

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, train_size = 0.7, 
                                                   test_size = 0.3, random_state = 100)
```

```python
X2_train_sm = sm.add_constant(X2_train)
```

```python
lr = sm.OLS(y2_train, X2_train_sm).fit()
lr.params
```

```python
lr.summary()
```

```python
plt.scatter(X2_train, y2_train)
plt.plot(X_train, -76228.866347 + 491.400522*X_train,'r')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, train_size = 0.7, 
                                                   test_size = 0.3, random_state = 100)
```

```python
X3_train_sm = sm.add_constant(X3_train)
```

```python
lr = sm.OLS(y3_train, X3_train_sm).fit()
lr.params
```

## Error Terms

Finally, plot your error terms!

```python
import warnings
warnings.filterwarnings('ignore')

y_train_pred = lr.predict(X_train_sm)
res = (y_train - y_train_pred)

fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.x_label('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
plt.scatter(X_train,res)
plt.show()
```

```python
y2_train_pred = lr.predict(X2_train_sm)
res = (y2_train - y2_train_pred)

fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.x_label('y2_train - y2_train_pred', fontsize = 15)
plt.show()
```

```python
plt.scatter(X2_train,res)
plt.show()
```

```python
y3_train_pred = lr.predict(X3_train_sm)
res = (y3_train - y3_train_pred)

fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.x_label('y3_train - y3_train_pred', fontsize = 15)
plt.show()
```

```python
plt.scatter(X3_train,res)
plt.show()
```
